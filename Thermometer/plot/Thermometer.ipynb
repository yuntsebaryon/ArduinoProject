```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
```

```python
fName = '/Users/yuntse/data/arduino/TempSensor/TempSensor_20221107.cvs'
```

```python
Temp = pd.read_csv(fName, header = None, skiprows = 1, names = ['Time', 'TC_Temp', 'CJT_Temp', 
                                                                'RTD_Value', 'RTD_Ratio', 'RTD_R', 'RTD_Temp' ])
```

```python
Temp['Time'] = Temp['Time'].transform(lambda t: datetime.strptime(t, '%Y-%m-%d %H:%M:%S'))
```

```python
# create figure and axis objects with subplots()
fig, ax = plt.subplots(figsize = (20,8), layout = 'tight')

# make a plot
ax.plot( Temp['Time'], Temp['TC_Temp'], label = 'Thermocouple', lw = 3)
ax.plot( Temp['Time'], Temp['RTD_Temp'], label = 'RTD', lw = 3)
# set x-axis label
ax.set_xlabel('Time', fontsize = 26)
ax.tick_params(axis = 'x', labelrotation = 45, labelsize = 26)
# set y-axis label
ax.set_ylabel('Temperature ['+u'\u00b0'+'C]', fontsize = 26)
ax.tick_params(axis = 'y', labelsize = 26)

# set plot title
# ax.set_title( label = '2022/11/7 Temperature in Hot Water', fontsize = 26)

ax.grid('.')
ax.legend(loc = 'best', ncol = 1, fontsize = 26)

plt.show()

# save the plot as a file
fig.savefig('/Users/yuntse/Documents/ArduinoProject/TempSensor/TempSensor_hotwater_20221107.png')
```
